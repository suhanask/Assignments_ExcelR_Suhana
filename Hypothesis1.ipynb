```python
import pandas as pd
from scipy import stats
import scipy
```

```python
cutlets = pd.read_csv('Cutlets.csv')
cutlets
```

```python
cutlets.info()
```

```python
cutlets.describe()
```

```python
unitA = 'Unit A'
cutlet_A = cutlets[[unitA]]
```

```python
cutlet_A
```

```python
unitB = 'Unit B'
cutlet_B = cutlets[[unitB]]
```

```python
cutlet_B
```

```python
# null hypothesis = no difference in diameters of unit A and unit B
# alternate hypothesis = difference in diameters of unit A and unit B
# significance level = 5%
```

if p-value is less than 0.05 then we accpet null hypothesis.

```python
stats.ttest_ind(cutlet_A , cutlet_B)
```

Since p-value is more than 0.05, we accept the null hypothesis.
Hence, there is no significant difference in diameters of both unit A and B.
