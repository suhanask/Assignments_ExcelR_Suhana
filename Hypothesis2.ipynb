```python
import scipy
from scipy import stats
import pandas as pd
```

null hypothesis- no difference in average TAT among the labs


alternate hypothesis- there is difference in average TAT among the labs

```python
df = pd.read_csv('LabTAT.csv')
df
```

```python
df.isnull()
```

```python
df.describe()
```

```python
df.info()
```

```python
lab1 = 'Laboratory 1'
L1 = df[[lab1]]
```

```python
L1
```

```python
lab2 = 'Laboratory 2'
L2 = df[[lab2]]
lab3 = 'Laboratory 3'
L3 = df[[lab3]]
lab4 = 'Laboratory 4'
L4 = df[[lab4]]
```

```python
L2
```

```python
L3
```

```python
L4
```

performing ANOVA to compare all the four labs

```python
scipy.stats.f_oneway(L1, L2, L3, L4)
```

since p-value<0.05, we reject the null hypothesis.

Hence, there is a significant difference in average TAT among the labs.
