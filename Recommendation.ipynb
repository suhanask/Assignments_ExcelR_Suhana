```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import correlation,cosine
```

```python
book = pd.read_csv('book.csv', encoding='latin1')
```

```python
book
```

```python
data=book.drop('Unnamed: 0',axis=1)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data=data.rename(columns={'User.ID':'userid','Book.Title':'title','Book.Rating':'rating'})
data
```

```python
len(data.userid.unique())
```

```python
len(data.title.unique())
```

```python
books=data.pivot_table(index='userid',columns='title',values='rating').reset_index(drop=True)
books
```

```python
books.index=data.userid.unique()
books
```

```python
books.columns=data.title.unique()
books
```

```python
books.fillna(0,inplace=True)
books
```

```python
user_sim=1-pairwise_distances(books.values,metric='cosine')
user_sim
```

```python
np.fill_diagonal(user_sim,0)
user_sim
```

```python
user_df=pd.DataFrame(user_sim)
user_df
```
